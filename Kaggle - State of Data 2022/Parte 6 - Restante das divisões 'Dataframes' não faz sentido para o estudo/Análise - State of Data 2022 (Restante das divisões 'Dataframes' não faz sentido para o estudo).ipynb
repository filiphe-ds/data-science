## Metodologia

### Configuração inicial

```python
#Importando bibliotecas utilizadas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Configurando Pandas para exibir todas as colunas (são muitas colunas)
pd.set_option('display.max_columns', None)

#Revertendo a configuração
#pd.reset_option('display.max_columns')
```

```python
#Carregamento do Dataset

data = pd.read_csv('data/State_of_data_2022.csv')
```

```python
#Melhorando a leitura das colunas

data.columns = [coluna.strip("()").replace("'", "").replace(", ", "_") for coluna in data.columns]
```

```python
#Testando o novo modo de leitura das colunas
data.head()
```

```python
#Acessando o id dos usuários pesquisados para criar novos datasets, separando todas as 8 partes da pesquisa
#Deixando-as vinculadas ao id

coluna_principal = data.iloc[:,0]
```

### Trabalhando os dados de análise de objetivos

```python
#Separando todas as perguntas da parte 6 em um novo dataset para análise de conhecimento de DE (Engenharia de dados)

colunas_p6 = [coluna for coluna in data.columns if coluna.startswith('P6')]

data_parte6 = pd.concat([coluna_principal] + [data[coluna] for coluna in colunas_p6], axis=1)

data_parte6.head()
```

```python
#Separando todas as perguntas da parte 7 em um novo dataset para análise de conhecimento de DA (Análise de dados)

colunas_p7 = [coluna for coluna in data.columns if coluna.startswith('P7')]

data_parte7 = pd.concat([coluna_principal] + [data[coluna] for coluna in colunas_p7], axis=1)

data_parte7.head()
```

```python
#Separando todas as perguntas da parte 8 em um novo dataset para análise de conhecimento em DS (Ciência de dados)

colunas_p8 = [coluna for coluna in data.columns if coluna.startswith('P8')]

data_parte8 = pd.concat([coluna_principal] + [data[coluna] for coluna in colunas_p8], axis=1)

data_parte8.head()
```
